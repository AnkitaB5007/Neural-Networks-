````python
# import necessary modules
import torch
import numpy as np # for data preparation
from torch import matmul, sigmoid
from torch.nn import MSELoss # Mean squared error

torch.manual_seed(7)

# number of epochs, learning rate and objective function should be the same for everyone
epochs = 1000
lr = 0.5
criterion = MSELoss()
````

a- Weights if initialized to all zeros or all ones, then the derivative of loss function w.r.t is the same for every w in W[1]. Biases ,have when initialized to 1's shift the activation function by 1 unit(analogous to a constant in a linear function). If weights are randomly initialized, it is certainly better than above cases. But one thing is when they are mapped to very large values or very low values, then after applying sigmoid , they get mapped to nearly 1's or zeroes. b- If there’s a single input to an operation that requires gradient, its output will also require gradient. Conversely, only if all inputs don’t require gradient, the output also won’t require it. weights(Wi, Wo) and biases(ba,bo) need gradient tracking. we can toggle it using requires_grad = True when defining the weights and bias variables as tensors.

````python
X = torch.tensor([[0,0],
                  [0,1],
                  [1,0],
                  [1,1]
                 ], dtype=torch.float32)
print('X', X)
Y = torch.tensor([[0],[1],[1],[0]], dtype=torch.float32).T
print('Y', Y)

# params
n_input, n_hidden, n_output = 2, 2, 1

W_i = torch.rand(n_input, n_hidden, dtype=torch.float32, requires_grad = True)
print('W_i', W_i)
W_o = torch.rand(n_hidden, n_output, dtype=torch.float32,requires_grad = True)
print('W_o', W_o)

b_a = torch.rand((1, n_hidden), requires_grad = True)
b_o = torch.rand((1, n_output), requires_grad = True)
print('b_a', b_a)
print('b_o', b_o)
````

## 2

The next step is to implement a full forward pass of the model. This means you have to implement the computation graph built in a) and present each XOR input once to it.

````python
# TODO: forward pass
def forward(x, W_i, b_a, W_o, b_o):
    z2 = torch.matmul(X, W_i) + b_a
    # activation of hidden layer 
    a2 = torch.sigmoid(z2)
    #a2 = a2.type(torch.long)
    # activation (output) of final layer 
    z2 = torch.matmul(a2, W_o) + b_o
    Y_hat =  torch.sigmoid(z2)
    print(Y_hat)
    return Y_hat

Y_hat = forward(X,W_i,b_a,W_o,b_o)

#loss1 = -torch.mean(Y*torch.log(Y_hat)+(1-Y)*torch.log(1-Y_hat))

loss = criterion(Y, Y_hat) # Loss function
print(loss)


````

## 3
Enhance your model by actually doing backpropagation! 

Gradient from W_o, W_i + biases

a) Employ the criterion defined in the first code cell to get a loss value and backpropagate by using the .backward() method. Then, access the gradient information in the tracked variables and perform the update. You can do so for a variable w by:

```python
w.data -= lr*w.grad.data
```

You also need to reset the gradient data after performing the update:

```python
w.grad.data.zero_()
```

You will have to loop over the data set several times to get a nice outcome. If your implementation is correct the values defined above (number of epochs, learning rate) will suffice.

````python

for epoch in range(epochs):
    #TODO: Forward pass
    #TODO: Backward pass
    Y_hat = forward(X,W_i,b_a,W_o,b_o)
    loss = criterion(Y, Y_hat) # Loss function
    print(loss)
    #backward propagation
    loss.requires_grad_(True)
    loss.backward()
    
    # optimise weights    
    with torch.no_grad():
        W_i -= W_i.grad.data * lr
        W_o -= W_o.grad.data * lr
        b_a -= b_a.grad.data * lr
        b_o -= b_o.grad.data * lr
        W_i.grad.zero_()
        W_o.grad.zero_()
        b_a.grad.zero_()
        b_o.grad.zero_()
    H_para = [W_i,W_o,b_a,b_o]
#print the hyperparameters that is weights and biases 
# W_i,b_a,W_o,b_o(scalar) respectively
print('W_i',H_para[0]) 
print('b_a',H_para[1])
print('W_o',H_para[2])
print('b_o',H_para[3])
````
